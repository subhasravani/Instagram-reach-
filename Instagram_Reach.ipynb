## Feature engineering

### Subtask:
Prepare the data for model training by performing feature engineering.


**Reasoning**:
One-hot encode the 'USERNAME' column, scale numerical features, and combine them into a final dataframe. Handle potential errors gracefully.



```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

try:
    # 1. One-hot encode 'USERNAME'
    df_encoded = pd.get_dummies(df, columns=['USERNAME'], drop_first=True)

    # 2. Scale numerical features
    numerical_features = ['Followers', 'Likes', 'Hashtag Count', 'Day of Week', 'Hour of Day']
    scaler = StandardScaler()
    df_encoded[numerical_features] = scaler.fit_transform(df_encoded[numerical_features])

    scaled_columns = [f"{col}_scaled" for col in numerical_features]
    df_encoded[scaled_columns] = scaler.fit_transform(df_encoded[numerical_features])

    # 3. Combine features and drop original columns
    columns_to_drop = ['USERNAME_Other', 'Followers', 'Likes', 'Hashtag Count', 'Day of Week', 'Hour of Day']
    df_final = df_encoded.drop(columns=columns_to_drop, errors='ignore')

    display(df_final.head())

except KeyError as e:
    print(f"Error: Key error encountered - {e}.  Check if the specified column exists.")
except ValueError as e:
    print(f"Error: Value error encountered - {e}.  Check the data types and values.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
